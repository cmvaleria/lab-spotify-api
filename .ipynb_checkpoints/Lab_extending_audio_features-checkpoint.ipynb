```python
import time
```

creating the function search_song(title, artist):

```python
def search_song(title, artist):
    time.sleep(5)
    try:
        results = sp.search(q=title+','+artist, limit=5)
        return results['tracks']['items'][-1]['id']
    except: 
        return 'na'
    
```

### 2. Create a function get_audio_features(list_of_songs) to obtain the audio features of a given list of songs (the content of list_of_songs can be the href/id/uri).

```python
def get_audio_features(list_of_songs):
    my_dict = {}
    los=[]
    for i in list_of_songs:
        time.sleep(5)
        try:
            x = sp.audio_features(i)[0]
            if x: los.append(x)
        except:
            pass
    df = pd.DataFrame(los)
    return df
```

```python
not_hot_songs1_ids = pd.read_csv('files_exported/not_hot_songs1_ids.csv')
not_hot_songs2_ids = pd.read_csv('files_exported/not_hot_songs1_ids.csv')
not_hot_songs3_ids = pd.read_csv('files_exported/not_hot_songs3_ids.csv')
not_hot_songs_ids_final = pd.concat([not_hot_songs1_ids,not_hot_songs2_ids,not_hot_songs3_ids]).reset_index()
not_hot_songs_ids_final = not_hot_songs_ids_final.drop(['index'], axis=1)
not_hot_songs_ids_final.to_csv('files_exported/not_hot_songs_ids_final.csv',index=False)
```

```python
not_hot_songs1_feat = pd.read_csv('files_exported/not_hot_songs1_feat.csv')
not_hot_songs2_feat = pd.read_csv('files_exported/not_hot_songs2_feat.csv')
not_hot_songs3_feat = pd.read_csv('files_exported/not_hot_songs3_features.csv')
not_hot_songs_feat_final = pd.concat([not_hot_songs1_feat,not_hot_songs2_feat,not_hot_songs3_feat]).reset_index()
not_hot_songs_feat_final = not_hot_songs_feat_final.drop(['index'], axis=1)
not_hot_songs_feat_final.to_csv('files_exported/not_hot_songs_feat_final.csv',index=False)
```

```python
display(not_hot_songs_feat_final.tail())
#not_hot_songs_feat_final.shape
```

Now the function to concatenate:

```python
def add_audio_features(df, audio_features_df):
    df['check'] = df.id.isin(audio_features_df.id)
    df = df[(df['check'] == True)]
    df.reset_index(drop=True, inplace=True)
    df = df.drop(['check'], axis=1)
    x = str(input('How do you want to call the dataframe? Make sure to provide file extension:\n'))
    y = 'files_exported/'
    extended_df = pd.concat([df, audio_features_df], axis=1)
    extended_df.to_csv((y+x), index=False)
    return extended_df
```
